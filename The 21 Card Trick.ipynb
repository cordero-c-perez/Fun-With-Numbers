## <center>The 21 Card Trick</center>

Basically, the trick is the following: (learn more [here](https://en.wikipedia.org/wiki/Twenty-One_Card_Trick))

1. Choose 21 random cards from the deck and ask a volunteer to memorize 1
2. Lay them across 3 piles (7x3) grid
3. Ask the subject to identify the pile in which thier memorized card resides
4. Place that pile in the middle, flip the card stack, repeat steps 2-4 (2x)
5. Flip over the 11th card and that is the volunteer's card

Running the following cell will start a game in which you are the volunteer. The final matrix is displayed purely to show that by following the steps the memorized card will always move to the center of the middle pile, and hence why it is the 11th card. We can generalize this to be: rows + (rows % columns) = memorized card.

In a 7x3 grid. The memorized card will move to the middle position which is 7 + (7%3) = 11.

###### What happens when there are more than 3 columns?

The previous function will not work as the middle pile now has more than 1 column of values before and after it. 

To generalize further, the memorized card will always be in the nth postion where 
n = floor(columns/2) + (rows % columns). To see an example of this, run the bottom-most cell as it is an expanded version of the 21 card trick. This version uses odd-valued rows and columns. The code can be changed to include even values and the formula for the position of the memorized card can then be modified accordingly.
